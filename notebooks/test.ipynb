```python
import sys
import importlib
import json

# Force re-importing of local modules by clearing from sys.modules
for module_name in list(sys.modules.keys()):
    if module_name.startswith("src.") or module_name == "src":
        sys.modules.pop(module_name, None)

from src.pipeline.resume_parser import parse_resume
```

```python
# Path to the resume (can be a remote Google Drive URL, GitHub URL, or a local path)
resume_file_path = "data/Resume.pdf"

# Run unified parsing & embedding pipeline (Option B)
parsed_response = parse_resume(resume_file_path, include_embedding=True)

# Save full parsed response (including embeddings and metadata) as a JSON file
with open("data/resume_data.json", "w", encoding="utf-8") as f:
    json.dump(parsed_response, f, ensure_ascii=False, indent=4)

print("Resume parsed and embedding generated successfully!")
print("Keys in response payload:", list(parsed_response.keys()))
```

### Embeddings & Metadata Analysis

Inspect the canonical embedding text, vector metadata, and generated 384-dimensional embedding vector.

```python
# 1. Canonical Embedding Text (Notice PII & CGPA are excluded)
print("=== Canonical Embedding Text ===")
print(parsed_response["embedding_text"])

print("\n=== Embedding Metadata ===")
print(json.dumps(parsed_response["embedding_metadata"], indent=2))

# 2. Vector Embedding Output
embedding_vector = parsed_response["embedding"]
if embedding_vector:
    print(f"\n=== Embedding Vector ===")
    print(f"Vector Dimension: {len(embedding_vector)}")
    print(f"First 5 Vector Values: {embedding_vector[:5]}")
else:
    print("\nEmbedding generation returned null (check embedding_metadata for details).")
```
